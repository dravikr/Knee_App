```python
!pip install ultralytics --upgrade
from ultralytics import YOLO

yaml_content = """
train: /content/drive/MyDrive/knee_data/images/train
val: /content/drive/MyDrive/knee_data/images/val
nc: 5
names: ['Healthy', 'Doubtful', 'Minimal', 'Moderate', 'Severe']
"""

yaml_path = "/content/drive/MyDrive/knee_data/data.yaml"
with open(yaml_path, "w") as f:
    f.write(yaml_content)

model = YOLO('yolov8m.pt')
model.train(
    data=yaml_path,
    epochs=120,
    imgsz=640,
    batch=8,
    device=0,
    project='/content/drive/MyDrive/knee_data/runs',
    name='exp_yolov8m_gpu'
)
```
